```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
df
```

```python
df.describe()
```

```python
plt.plot(df['Hours'],df['Scores'],"o",markersize=2)
```

```python
import statsmodels.formula.api as smf

line_info = smf.ols('Scores ~ Hours',data=df).fit() # this line fits the data to get the line of least squares "to get the best weights" cooeffecient of my linear regression
pred = line_info.predict(df)  # predict the 'Scores' based on my data

plt.plot(df['Hours'],df['Scores'],"o",markersize=2)
plt.plot(df['Hours'],pred,"-")
```
